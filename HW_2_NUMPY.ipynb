```python
a1, a2 = np.vsplit(a, 2)

print("Перша частина масиву:")
print(a1)

print("Друга частина масиву:")
print(a2)
```

```python
a1_sorted = a1[a1[:, 0].argsort()]

a2_sorted = a2[a2[:, 0].argsort()[::-1]]
```

```python
print("Перша частина масиву:")
print(a1_sorted)

print("Друга частина масиву:")
print(a2_sorted)
```

```python
result = np.concatenate((a1_sorted, a2_sorted), axis=0)
```

```python
print(result)
```

```python
values2, counts2  = np.unique(result, return_counts=True)
most_frequent_value = values2[np.argmax(counts2)]
```

```python
print (most_frequent_value)
```

```python
def modify_column(column):
    column_mean = np.mean(column)
    return np.where(column < column_mean, column * 2, column / 4)
```

```python
result[:, 2] = modify_column(result[:, 2])
```
